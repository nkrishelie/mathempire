#### Некоторые графики для конспекта

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import matplotlib
plt.rcParams.update({'font.size':20})
matplotlib.rcParams['text.usetex'] = True
from matplotlib import rc
#rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
#rc('text', usetex=True)
```

```python
def f(n,m):
    return m/(1+(n-m)**2)
```

```python
n = np.arange(0,10,0.01)
```

```python
plt.figure(figsize=(9,5))
for m in range(1,6):
    y = f(n,m)
    plt.plot(n,y, label='m=%s'%m)
plt.legend(loc='best')
plt.xlabel('$$n$$', fontsize=18)
plt.ylabel('$$x_{n,m}$$', fontsize=18)
#plt.suptitle(r'$\displaystyle y=\frac{m}{(1+(n-m)^2}$')
plt.savefig('limits.png')
```

```python
def h(x):
    return x/(1-x) if x>=0 else x/(1+x)
```

```python
x = np.arange(-0.95,0.96,0.01)
plt.figure(figsize=(5,6))
y=[h(z) for z in x]
plt.axvline(0,color='r', linewidth='0.5')
plt.axhline(0, color='black', linewidth='0.5')
plt.plot(x,y)
#plt.legend(loc='best')
plt.savefig('interval.png')
```

```python
z = 3+4j
e = math.exp(z.real)*(math.cos(z.imag)+math.sin(z.imag)*1j)
n = np.arange(0,100,1)
Z = np.array([(z**k)/math.factorial(k) for k in n])
E=[]
for f in Z:
    E.append(E[-1]+f) if len(E)>0 else E.append(1)
E = np.array(E)
plt.figure(figsize=(11,10))
plt.axvline(0,color='black', linewidth='0.5')
plt.axhline(0, color='black', linewidth='0.5')
plt.axvline(e.real,color='blue', linewidth='1')
plt.axhline(e.imag, color='blue', linewidth='1')
plt.plot(E.real,E.imag, color='red')
#plt.suptitle("exp%s = %s" % (z,e))
plt.savefig('exp-series.png')
```

### Корень

```python
x = np.arange(0,2,0.001)
n = [1,2,3,5,7,11,53]
```

```python
plt.figure(figsize=(11,5))
plt.axvline(0,color='black', linewidth='0.5')
plt.axvline(1,color='r', linewidth='0.5')
plt.axhline(0, color='black', linewidth='0.5')
plt.axhline(1, color='r', linewidth='0.5')
for param in n:
    y = [t**(1/param) for t in x]
    plt.plot(x,y, label="n=%s" % param)
plt.legend(loc='lower right')
plt.suptitle("$\displaystyle y=\sqrt[n]{x}$")
plt.savefig('root.png')
```

### Факториальные последовательности

```python
a = [1.2,1.3,1.4,1.5]
k = [5,6,6,7]
n = np.arange(0,100,1)
```

```python
plt.figure(figsize=(11,5))
#plt.axvline(0,color='black', linewidth='0.5')
#plt.axvline(1,color='r', linewidth='0.5')
#plt.axhline(0, color='black', linewidth='0.5')
#plt.axhline(1, color='r', linewidth='0.5')
for param,j in zip(a,k):
    y = [t**j/param**t for t in n]
    plt.plot(n,y, label="a=%s, k=%s" % (param,j))
plt.legend(loc='upper right')
plt.xlabel('n')
plt.suptitle(r"$\displaystyle y=n^k/a^n$")
plt.savefig('power.png')
```

```python
x = [5,6,7,8]
n = np.arange(0,30,1)
```

```python
plt.figure(figsize=(11,5))
#plt.axvline(0,color='black', linewidth='0.5')
#plt.axvline(1,color='r', linewidth='0.5')
#plt.axhline(0, color='black', linewidth='0.5')
#plt.axhline(1, color='r', linewidth='0.5')
for param in x:
    y = [param**t/math.factorial(t) for t in n]
    plt.plot(n,y, label="x=%s" % param)
plt.legend(loc='upper right')
plt.xlabel('n')
plt.suptitle(r'$\displaystyle y=x^n/n!$')
plt.savefig('fact.png')
```

#### Производная

```python
from sympy import *
#init_printing()
z = Symbol('x')
f = exp(z)
x0= 0
delta = [-2,-1,1,2]
g = diff(f)
x = np.arange(-2,4,0.01)
print_latex("f'(x)=%s" % latex(g.subs(z,0.5)))
```

```python
(f.subs(z,x0+1)-f.subs(z,x0))/2*(z-x0)
```

```python
plt.figure(figsize=(13,6))
plt.xlim(-2, 4.5)
plt.ylim(0, 3)
plt.axvline(0,color='black', linewidth='0.5')
plt.axvline(-1,color='black', linewidth='0.5')
plt.axvline(1, color='black', linewidth='0.5')
plt.axvline(2, color='black', linewidth='0.5')
plt.axhline(1, color='black', linewidth='0.5')
y = [f.subs(z,t) for t in x]
plt.plot(x,y, label="$y=%s$" % latex(f), linewidth='2')
for d in delta:
    y = [(f.subs(z,x0+d)-f.subs(z,x0))*(t-x0)/d+f.subs(z,x0) for t in x]
    plt.plot(x,y, label="$y=%s,\; \Delta x=%s$" %
             (latex(((f.subs(z,x0+d)-f.subs(z,x0))/d*(z-x0)+f.subs(z,x0))), d),
             linewidth='0.5')
y = [g.subs(z,x0)*(t-x0)+f.subs(z,x0) for t in x]
plt.plot(x,y, label="$y=%s$" % latex(g.subs(z,x0)*(z-x0)+f.subs(z,x0)), linewidth='1.5')
plt.legend(loc='best')
plt.savefig('deriv.png')
```
